```python
# Shim verification: legacy names are class-identical re-exports, the legacy
# __all__ surface is intact, and the fracture changed exactly one thing about
# the abstract set (execute left it).
import cjm_substrate.core.interface as _iface
from cjm_substrate.core import capability as _cap
from cjm_substrate.core import wire as _wire

assert _iface.PluginInterface is _cap.ToolCapability
assert _iface.FileBackedDTO is _wire.FileBackedDTO

_legacy_surface = ['RELOAD_TRIGGER', 'WORKER_ENV_TEMPLATE_PLACEHOLDERS',
                   'FileBackedDTO', 'ConfigOption', 'FieldOptions', 'EnvVarSpec',
                   'expand_worker_env_template', 'template_check_placeholders',
                   'PluginInterface', 'capability_action', 'collect_capability_actions']
for _n in _legacy_surface:
    assert hasattr(_iface, _n), f"shim lost legacy name: {_n}"
    assert _n in _iface.__all__, f"shim __all__ lost legacy name: {_n}"

_ams = set(_iface.PluginInterface.__abstractmethods__)
assert 'execute' not in _ams, "execute must have left the abstract set"
assert _ams == {'name', 'version', 'initialize', 'get_config_schema',
                'get_current_config'}, _ams
print('interface shim verification OK')
```
